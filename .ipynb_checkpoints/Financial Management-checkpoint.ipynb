```python
from flask import Flask, render_template, request
import requests
import speech_recognition as sr
import csv

app = Flask(__name__)

# In-memory user database
users = [
    {'username': 'user1', 'password': 'password1'},
    {'username': 'user2', 'password': 'password2'},
    # Add more users as needed
]

# Serve login form
@app.route('/login', methods=['GET'])
def login_form():
    return render_template('login.html')

# Process login form submission
@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']
    password = request.form['password']

    # Check if the provided username and password match a user in the database
    user = next((user for user in users if user['username'] == username and user['password'] == password), None)
    if user:
        # Successful login
        return f"Welcome, {username}!"
    else:
        # Failed login
        return "Invalid username or password."


# OCR.space API endpoint
api_url = 'https://api.ocr.space/parse/image'
api_key = 'K83923397588957'

@app.route('/', methods=['GET', 'POST'])
def upload_image():
    if request.method == 'POST':
        # Get the uploaded file from the request
        file = request.files['image']

        # Send a POST request to the OCR.space API
        response = requests.post(api_url, files={'file': file}, data={'apikey': api_key})

        # Parse the response JSON and extract the recognized text
        if response.status_code == 200:
            result = response.json()
            if result['IsErroredOnProcessing']:
                return 'Error occurred during OCR processing'
            else:
                parsed_text = result['ParsedResults'][0]['ParsedText']
                save_data(parsed_text)  # Save the recognized text to a CSV file
                return parsed_text
        else:
            return 'Error: ' + str(response.status_code)

    return render_template('upload.html')


@app.route('/voice', methods=['POST'])
def voice_recognition():
    # Get the audio file from the request
    audio = request.files['audio']

    # Initialize the SpeechRecognition recognizer
    recognizer = sr.Recognizer()

    # Load the audio file
    with sr.AudioFile(audio) as source:
        audio_data = recognizer.record(source)

    # Perform speech recognition
    try:
        recognized_text = recognizer.recognize_google(audio_data)
        save_data(recognized_text)  # Save the recognized text to a CSV file
        return f"Recognized text: {recognized_text}"
    except sr.UnknownValueError:
        return "Speech recognition could not understand audio"
    except sr.RequestError as e:
        return f"Error occurred during speech recognition: {e}"


def save_data(text):
    with open('recognized_text.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([text])


if __name__ == '__main__':
    app.run(debug=True)
```
